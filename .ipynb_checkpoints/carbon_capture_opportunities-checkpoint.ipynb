```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Importing the dataset
file = "Databases\ghg_CO2_2010_2020_3D.csv"
emissions_df = pd.read_csv(file, low_memory=False, thousands=',')
emissions_df.head(5)
```

```python
emissions_df.dtypes
```

```python
# introducing the cost of carbon capture per cubic tons per industry (using NAICS codes with 3 digits)

# create a list of our conditions
conditions = [
    (emissions_df["naics_3digits"] == 211) |  (emissions_df["naics_3digits"] == 325),
   (emissions_df["naics_3digits"].isin([212, 238, 324, 327, 331, 332])),
   (emissions_df["naics_3digits"] == 221) & (emissions_df["fuel_type"] == "Coal"),
   (emissions_df["naics_3digits"] == 221) & (emissions_df["fuel_type"] == "Natural Gas")]

# create a list of the values we want to assign for each condition
values = ['30', '150', '65', '130']

# create a new column and use np.select to assign values to it using our lists as arguments
emissions_df["capture_cost"] = np.select(conditions, values)

# display updated DataFrame
emissions_df.head()
    
```

```python
emissions_df["capture_cost"].value_counts()[0]
```

```python
len(emissions_df["capture_cost"])
```

```python
# remove rows where the capture's cost is unknown by filtering
values = ['30', '150', '65', '130']
emissions_def_df = emissions_df[emissions_df["capture_cost"].isin(values) == True]
len(emissions_def_df)
```

```python
emissions_def_df["capture_cost"] = pd.to_numeric(emissions_df["capture_cost"])
```

```python
# calculating the cost of carbon capture for the 2020 emissions per plant
emissions_def_df["total_capture_cost"] = emissions_def_df["capture_cost"]*emissions_def_df["co2_emissions_2020"]
len(emissions_def_df["total_capture_cost"])
```

```python
# calculating the incentive per plant if the CO2 2020 emissions are captured 
# (with a tax credit of $50 per ton of carbon captured)
emissions_def_df["45Q_credit_incentive"] = emissions_def_df["co2_emissions_2020"]*50
len(emissions_def_df["45Q_credit_incentive"])
```

```python
# calculating the economic opportunity for the plant to invest in carbon capture
emissions_def_df["eco_opportunity"] = emissions_def_df["total_capture_cost"] - emissions_def_df["45Q_credit_incentive"]
len(emissions_def_df["eco_opportunity"])
```

```python
# qualifying the opprotunity


# create a list of our opportunity levels
opportunity = [
    (emissions_def_df["eco_opportunity"] < 10000000) & (emissions_def_df["co2_emissions_2020"]<500000),
    (emissions_def_df["eco_opportunity"] < 10000000) & (emissions_def_df["co2_emissions_2020"]>=500000),
    (emissions_def_df["eco_opportunity"] >= 10000000) & (emissions_def_df["co2_emissions_2020"]<500000),
    (emissions_def_df["eco_opportunity"] >= 10000000) & (emissions_def_df["co2_emissions_2020"]>=500000)]

# create a list of the values we want to assign for each opportunity level
values = ['Forget it', 'Too early', 'Too discuss', 'To target']

# create a new column and use np.select to assign values to it using our lists as arguments
emissions_def_df["business_opportunity"] = np.select(opportunity, values)

# display updated DataFrame
emissions_def_df.head()
```

```python
plt.scatter(emissions_def_df["co2_emissions_2020"], emissions_def_df["eco_opportunity"])
plt.show()
```

```python
emissions_def_df.to_csv("emissions_def.csv", index = False, header = True)
```
