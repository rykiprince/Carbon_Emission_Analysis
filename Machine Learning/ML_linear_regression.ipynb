```python
import pandas as pd
import numpy as np
from pathlib import Path
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn import linear_model

from sklearn import tree
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
```

```python
ML_df = pd.read_csv(Path(r'C:\Users\helen\Desktop\Berkeley boot camp\Datasets\carbon_emission_analysis\Jupyter notebook/ml_database.csv'))
ML_df.head()
```

```python
# Select the columns we need for the linear regression for the year 2020
ML_2020_df = ML_df.drop(["Facility Id"], axis = 1).loc[(ML_df["year"] == 2020)]
```

```python
# Transform the categorical features into numerical data
from sklearn import preprocessing

le = preprocessing.LabelEncoder()
ML_2020_df['State'] = le.fit_transform(ML_2020_df['State'])
ML_2020_df['industry_type'] = le.fit_transform(ML_2020_df['industry_type'])
ML_2020_df.head()
```

```python
# Define the features set
X = ML_2020_df.copy()
X = X.drop("total_direct_emissions", axis=1)
X.head()
```

```python
# Define the target set.
y = ML_2020_df["total_direct_emissions"].values
y[:5]
```

```python
# Split the Data Into Training and Testing Sets into an 80/20 split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1, train_size=0.80)
```

```python
# Scale the Training and Testing Data
scaler = StandardScaler()

# Fit and scale the Standard Scaler with the training data
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.fit_transform(X_test)
```

```python
#  Create the decision tree regressor instance
model = DecisionTreeRegressor(random_state=0)

# Fit Model
model = model.fit(X_train_scaled, y_train)
```

```python
# Make predictions using the testing data
predictions = model.predict(X_test_scaled)
predictions
```

```python
# Test the balanced accuracy score of the model
from sklearn.metrics import r2_score
r2_score(y_test, predictions)
```
