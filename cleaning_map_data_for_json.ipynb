```python
from collections import OrderedDict
import json
li=[]
for index,row in df.iterrows():
    d = OrderedDict()
    d['type']='Feature'
    d['geometry']={
        'type':'Point',
        'coordinates':[row['longitude'],row['latitude']]
    }
    d['properties']={
        'carbon_dioxide_emission':[row['carbon_dioxide_emission']]
    }
    li.append(d)
    
d=OrderedDict()
d['type']='FeatureCollection'
d['features']=li
# print(li[2]['geometry'])
with open('clean_map_data_full.json','w') as f:
    f.write(json.dumps(d,sort_keys=False, indent=4))
```
