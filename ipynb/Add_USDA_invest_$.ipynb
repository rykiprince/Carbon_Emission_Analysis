```python
import pandas as pd
import numpy as np
import os
```

```python
# Load USDA number of investments data
USDA_USDinvest_data = os.path.join(r'C:\Users\helen\Desktop\Berkeley boot camp\Datasets\carbon_emission_analysis\Historical data\USDA_Energy_$I_2016-2020.csv')
USDA_USDinvest_df = pd.read_csv(USDA_USDinvest_data)
USDA_USDinvest_df.head()
```

```python
# Load Emissions data
emission_data = os.path.join(r'C:\Users\helen\Desktop\Berkeley boot camp\Datasets\carbon_emission_analysis\Jupyter notebook\ml_database.csv')
emission_df = pd.read_csv(emission_data)
emission_df.head()
```

```python
USDA_USDinvest_2016 = USDA_USDinvest_df[["State abbreviation","2016"]]
USDA_USDinvest_2016["year"]=2016
USDA_USDinvest_2016["USDA_number_invest"] = USDA_USDinvest_2016.pop("2016")

USDA_USDinvest_2017 = USDA_USDinvest_df[["State abbreviation","2017"]]
USDA_USDinvest_2017["year"]=2017
USDA_USDinvest_2017["USDA_number_invest"] = USDA_USDinvest_2017.pop("2017")

USDA_USDinvest_2018 = USDA_USDinvest_df[["State abbreviation","2018"]]
USDA_USDinvest_2018["year"]=2018
USDA_USDinvest_2018["USDA_number_invest"] = USDA_USDinvest_2018.pop("2018")

USDA_USDinvest_2019 = USDA_USDinvest_df[["State abbreviation","2019"]]
USDA_USDinvest_2019["year"]=2019
USDA_USDinvest_2019["USDA_number_invest"] = USDA_USDinvest_2019.pop("2019")

USDA_USDinvest_2020 = USDA_USDinvest_df[["State abbreviation","2020"]]
USDA_USDinvest_2020["year"]=2020
USDA_USDinvest_2020["USDA_number_invest"] = USDA_USDinvest_2020.pop("2020")
USDA_USDinvest_2020.head()
```

```python
USDA_USDinvest_df = pd.concat([USDA_USDinvest_2016,
                            USDA_USDinvest_2017,
                            USDA_USDinvest_2018,
                            USDA_USDinvest_2019,
                            USDA_USDinvest_2020
                           ],axis=0)
USDA_USDinvest_df.dropna()
USDA_USDinvest_df.tail()
```

```python
USDA_USDinvest_df = USDA_USDinvest_df.rename(columns = {"State abbreviation" : "State"})
```

```python
emission_df = pd.merge(emission_df, USDA_USDinvest_df, on=["State","year"])
emission_df
```

```python
emission_df.to_csv("ml_database.csv",index=False)
```
