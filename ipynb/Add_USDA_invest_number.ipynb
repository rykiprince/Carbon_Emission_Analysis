```python
import pandas as pd
import numpy as np
import os
```

```python
# Load USDA number of investments data
USDA_invest_data = os.path.join(r'C:\Users\helen\Desktop\Berkeley boot camp\Datasets\carbon_emission_analysis\Historical data\usda_Energy_number_I_2016-2020.csv')
USDA_invest_df = pd.read_csv(USDA_invest_data)
USDA_invest_df.head()
```

```python
# Load Emissions data
emission_data = os.path.join(r'C:\Users\helen\Desktop\Berkeley boot camp\Datasets\carbon_emission_analysis\DB_Resources\ml_database.csv')
emission_df = pd.read_csv(emission_data)
emission_df.head()
```

```python
USDA_invest_2016 = USDA_invest_df[["State abbreviations","2016"]]
USDA_invest_2016["year"]=2016
USDA_invest_2016["USDA_number_invest"] = USDA_invest_2016.pop("2016")

USDA_invest_2017 = USDA_invest_df[["State abbreviations","2017"]]
USDA_invest_2017["year"]=2017
USDA_invest_2017["USDA_number_invest"] = USDA_invest_2017.pop("2017")

USDA_invest_2018 = USDA_invest_df[["State abbreviations","2018"]]
USDA_invest_2018["year"]=2018
USDA_invest_2018["USDA_number_invest"] = USDA_invest_2018.pop("2018")

USDA_invest_2019 = USDA_invest_df[["State abbreviations","2019"]]
USDA_invest_2019["year"]=2019
USDA_invest_2019["USDA_number_invest"] = USDA_invest_2019.pop("2019")

USDA_invest_2020 = USDA_invest_df[["State abbreviations","2020"]]
USDA_invest_2020["year"]=2020
USDA_invest_2020["USDA_number_invest"] = USDA_invest_2020.pop("2020")
USDA_invest_2020.head()
```

```python
USDA_invest_df = pd.concat([USDA_invest_2016,
                            USDA_invest_2017,
                            USDA_invest_2018,
                            USDA_invest_2019,
                            USDA_invest_2020
                           ],axis=0)
USDA_invest_df.dropna()
USDA_invest_df.tail()
```

```python
USDA_invest_df = USDA_invest_df.rename(columns = {"State abbreviations" : "State"})
```

```python
emission_df = pd.merge(emission_df, USDA_invest_df, on=["State","year"])
emission_df
```

```python
emission_df.to_csv("ml_database.csv",index=False)
```
