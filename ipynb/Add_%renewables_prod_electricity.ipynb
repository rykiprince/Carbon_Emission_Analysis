```python
import pandas as pd
import numpy as np
import os
```

```python
# Load USDA number of investments data
renewables_data = os.path.join(r'C:\Users\helen\Desktop\Berkeley boot camp\Datasets\carbon_emission_analysis\Historical data\Renewable_Electricity_prod_2016-2020.csv')
renewables_df = pd.read_csv(renewables_data)
renewables_df.head()
```

```python
# Load Emissions data
emission_data = os.path.join(r'C:\Users\helen\Desktop\Berkeley boot camp\Datasets\carbon_emission_analysis\Jupyter notebook\ml_database.csv')
emission_df = pd.read_csv(emission_data)
emission_df.head()
```

```python
renewables_2016 = renewables_df[["State abbreviation","2016"]]
renewables_2016["year"]=2016
renewables_2016["%Renewables"] = renewables_2016.pop("2016")

renewables_2017 = renewables_df[["State abbreviation","2017"]]
renewables_2017["year"]=2017
renewables_2017["%Renewables"] = renewables_2017.pop("2017")

renewables_2018 = renewables_df[["State abbreviation","2018"]]
renewables_2018["year"]=2018
renewables_2018["%Renewables"] = renewables_2018.pop("2018")

renewables_2019 = renewables_df[["State abbreviation","2019"]]
renewables_2019["year"]=2019
renewables_2019["%Renewables"] = renewables_2019.pop("2019")

renewables_2020 = renewables_df[["State abbreviation","2020"]]
renewables_2020["year"]=2020
renewables_2020["%Renewables"] = renewables_2020.pop("2020")
renewables_2020.head()
```

```python
renewables_df = pd.concat([renewables_2016,
                            renewables_2017,
                            renewables_2018,
                            renewables_2019,
                            renewables_2020
                           ],axis=0)
renewables_df.dropna()
renewables_df.tail()
```

```python
renewables_df = renewables_df.rename(columns = {"State abbreviation" : "State"})
renewables_df
```

```python
emission_df = pd.merge(emission_df, renewables_df, on=["State","year"])

emission_df = emission_df.rename(columns = {"population" : "state_population",
                                           "gdp_per_capita" : "state_gdp_per_capita",
                                           "policies_incentives" : "state_policies_incentives",
                                           "USDA_number_invest_x" : "USDA_energy_invest_unit",
                                           "USDA_number_invest_y" : "USDA_energy_invest_$"})
emission_df
```

```python
emission_df.to_csv("ml_database.csv",index=False)
```
