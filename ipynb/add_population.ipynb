```python
import pandas as pd
import numpy as np
```

```python
# Load Populatiion data
population_data = pd.read_excel("Population_2016-2020.xls", sheet_name=0,header=5)
population_df = pd.DataFrame(population_data)
population_df.head()
```

```python
# Load Emissions data
emission_data = pd.read_csv("emission_by_year.csv")
emission_df = pd.DataFrame(emission_data)
emission_df.head()
```

```python
emission_df = emission_df.rename(columns={"reporting_year":"year"})
emission_df
```

```python
# Load State name data
state_name = pd.read_excel("state_abbr.xlsx", sheet_name=0, header = 0)
state_df = pd.DataFrame(state_name)
state_df.head()
```

```python
population_df["State"] = population_df["GeoName"].str.upper()
population_df.head()
```

```python
# Change the State name to abbr
population_df = pd.merge(population_df,state_df,on="State")
population_df = population_df[["Abbreviation", "2016","2017","2018","2019","2020"]]
population_df = population_df.rename(columns={"Abbreviation":"State"})
population_df.head()
```

```python
population_2016 = population_df[["State","2016"]]
population_2016["year"]=2016
population_2016["population"] = population_2016.pop("2016")

population_2017 = population_df[["State","2017"]]
population_2017["year"]=2017
population_2017["population"] = population_2017.pop("2017")
population_2017.head()

population_2018 = population_df[["State","2018"]]
population_2018["year"]=2018
population_2018["population"] = population_2018.pop("2018")

population_2019 = population_df[["State","2019"]]
population_2019["year"]=2019
population_2019["population"] = population_2019.pop("2019")

population_2020 = population_df[["State","2020"]]
population_2020["year"]=2020
population_2020["population"] = population_2020.pop("2020")
population_2020.head()
```

```python
population_df = pd.concat([population_2016,
                           population_2017,
                           population_2018,
                           population_2019,
                           population_2020
                           ],axis=0)
population_df.dropna()
population_df.head()
```

```python
emission_df = pd.merge(emission_df,population_df,on=["State","year"])
emission_df
```

```python
emission_df.drop(columns=['Latitude','Longitude','Primary NAICS Code'])
new_columns = ['Facility Id','State','industry_type','year','total_direct_emissions','population']
emission_df = emission_df[new_columns]
emission_df
```

```python
emission_df.to_csv("ml_database.csv",index=False)
```
