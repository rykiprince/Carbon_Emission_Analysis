```python
import pandas as pd
import numpy as np
```

```python
# Load Populatiion data
gdp_data = pd.read_excel("GDP_capita_2016-2020.xlsx", sheet_name=0,header=5)
gdp_df = pd.DataFrame(gdp_data)
gdp_df.head()
```

```python
# Load Emissions data
emission_data = pd.read_csv("ml_database.csv")
emission_df = pd.DataFrame(emission_data)
emission_df.head()
```

```python
# Load State name data
state_name = pd.read_excel("state_abbr.xlsx", sheet_name=0, header = 0)
state_df = pd.DataFrame(state_name)
state_df
```

```python
gdp_df["State"] = gdp_df["GeoName"].str.upper()
gdp_df
```

```python
# Change the State name to abbr
gdp_df = pd.merge(gdp_df,state_df,on="State")
gdp_df = gdp_df[["Abbreviation", "2016","2017","2018","2019","2020"]]
gdp_df = gdp_df.rename(columns={"Abbreviation":"State"})
gdp_df
```

```python
gdp_2016 = gdp_df[["State","2016"]]
gdp_2016["year"]=2016
gdp_2016["gdp_per_capita"] = gdp_2016.pop("2016")

gdp_2017 = gdp_df[["State","2017"]]
gdp_2017["year"]=2017
gdp_2017["gdp_per_capita"] = gdp_2017.pop("2017")
gdp_2017.head()

gdp_2018 = gdp_df[["State","2018"]]
gdp_2018["year"]=2018
gdp_2018["gdp_per_capita"] = gdp_2018.pop("2018")

gdp_2019 = gdp_df[["State","2019"]]
gdp_2019["year"]=2019
gdp_2019["gdp_per_capita"] = gdp_2019.pop("2019")

gdp_2020 = gdp_df[["State","2020"]]
gdp_2020["year"]=2020
gdp_2020["gdp_per_capita"] = gdp_2020.pop("2020")
gdp_2020.head()
```

```python
gdp_df = pd.concat([gdp_2016,
                           gdp_2017,
                           gdp_2018,
                           gdp_2019,
                           gdp_2020
                           ],axis=0)
gdp_df.dropna()
gdp_df.head()
```

```python
emission_df = pd.merge(emission_df,gdp_df,on=["State","year"])
emission_df
```

```python
emission_df.to_csv("ml_database.csv",index=False)
```
