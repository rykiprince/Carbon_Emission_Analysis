```python
import pandas as pd
import numpy as np
```

```python
# Load Populatiion data
incentive_data = pd.read_excel("Policies_incentives_2016-2020.xlsx", sheet_name=0,header=3)
incentive_df = pd.DataFrame(incentive_data)
incentive_df.head()
```

```python
# Load Emissions data
emission_data = pd.read_csv("ml_database.csv")
emission_df = pd.DataFrame(emission_data)
emission_df.head()
```

```python
# Load State name data
state_name = pd.read_excel("state_abbr.xlsx", sheet_name=0, header = 0)
state_df = pd.DataFrame(state_name)
state_df.head()
```

```python
incentive_df["State"] = incentive_df["State"].str.upper()
incentive_df.head()
```

```python
# Change the State name to abbr
incentive_df = pd.merge(incentive_df,state_df,on="State")
incentive_df.columns
```

```python
incentive_df = incentive_df.drop(columns=["State"])
incentive_df = incentive_df.rename(columns={"Abbreviation":"State"})
incentive_df.head()
```

```python
incentive_2016 = incentive_df[["State",2016]]
incentive_2016["policies_incentives"] = incentive_2016.pop(2016)
incentive_2016["year"]=2016

incentive_2017 = incentive_df[["State",2017]]
incentive_2017["policies_incentives"] = incentive_2017.pop(2017)
incentive_2017["year"]=2017

incentive_2018 = incentive_df[["State",2018]]
incentive_2018["policies_incentives"] = incentive_2018.pop(2018)
incentive_2018["year"]=2018

incentive_2019 = incentive_df[["State",2019]]
incentive_2019["policies_incentives"] = incentive_2019.pop(2019)
incentive_2019["year"]=2019

incentive_2020 = incentive_df[["State",2020]]
incentive_2020["policies_incentives"] = incentive_2020.pop(2020)
incentive_2020["year"]=2020
incentive_2020.head()
```

```python
incentive_df = pd.concat([incentive_2016,
                           incentive_2017,
                           incentive_2018,
                           incentive_2019,
                           incentive_2020
                           ],axis=0)
incentive_df
```

```python
emission_df = pd.merge(emission_df,incentive_df,on=["State","year"], how="outer")
emission_df.fillna(0,inplace=True)
emission_df
```

```python
emission_df.to_csv("ml_database.csv",index=False)
```
