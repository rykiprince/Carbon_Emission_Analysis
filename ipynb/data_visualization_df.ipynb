```python
emission_df['Primary NAICS Code'] = emission_df['Primary NAICS Code'].apply(str)
emission_df.dtypes
```

```python
emission_df["2017 NAICS code"] = emission_df['Primary NAICS Code'].str.extract('(\d{3})(?:\d{3})')
emission_df.head(10)
```

```python
NAICS_data = pd.read_excel("NAICS_code_list_3d.xlsx", sheet_name=0, header=0)
NAICS_df = pd.DataFrame(NAICS_data, columns=["2017 NAICS US   Code","2017 NAICS US Title"])
NAICS_df["2017 NAICS code"] = NAICS_df["2017 NAICS US   Code"].apply(str)
NAICS_df = NAICS_df[["2017 NAICS code","2017 NAICS US Title"]]
NAICS_df
```

```python
emission_by_industry = emission_df.merge(NAICS_df, how="left", on="2017 NAICS code")
emission_by_industry
```
